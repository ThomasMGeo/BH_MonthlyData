```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Have the change file name, no spaces!
filename = 'Rigs_by_State_012519.xlsx'
```

```python
df = pd.read_excel(filename, None) # kinda a big df
```

```python
xls = pd.ExcelFile(filename)
sheets = xls.sheet_names
type(sheets)
```

```python
sheets[0]
```

```python
# 13 is CA Land
# 16 is Colorado
# 63 is Utah
```

```python
# [Sheet][Column][Row]
df['Jan 19']['Unnamed: 9'][13]
```

```python
CA = []
CO = []
UT = []

for X in sheets:
    CA_data = df[X]['Unnamed: 9'][13]
    CA.append(CA_data)

    CO_data = df[X]['Unnamed: 9'][16]
    CO.append(CO_data)
    
    UT_data = df[X]['Unnamed: 9'][63]
    UT.append(UT_data)
```

```python
CA_array  = np.array(CA)
CO_array  = np.array(CO)
UT_array  = np.array(UT)
```

```python
len(sheets)
```

```python
plt.figure(figsize=(11,7))

plt.plot(sheets, CA_array, 'blue')
plt.plot(sheets, CO_array, 'c')
plt.plot(sheets, UT_array, 'green')

plt.ylim(0, 130)
plt.legend(('CA', 'CO', 'UT'))

plt.yticks(np.arange(0, 130, 10))
plt.xticks(np.arange(0, 230, 24)) # x label ever 2 years
plt.ylabel('Avg Number of Rigs', size=16)
plt.xlabel('Time', size=16)
plt.gca().invert_xaxis()

plt.savefig('MonthyChart.png', dpi=300)
```
